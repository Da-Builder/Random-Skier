```python
import subprocess
import os

# Just a script that (hopefully) should run after we set everything up on uni computer

# Clone the repo side by side
# /Projects
#   ├── my_analysis_tool/ (Your Repo)
#   │    └── run_pipeline.py
#   └── sam-body4d/       (Cloned separately)

# Run this script in a conda environment for sam-body4d

# Setup checkpoitns and config, run the things from sam-body4d root.






# Define paths relative to this script
BASE_DIR = os.path.dirname(os.path.abspath(__file__))
SAM_BODY4D_REPO_DIR = os.path.join(BASE_DIR, "..", "sam-body4d")

def run_sam_body4d(video_path, conda_env_path):
    # 1. Locate the python executable in your conda environment
    # Usually: /path/to/anaconda3/envs/body4d/bin/python
    python_path = os.path.join(conda_env_path, "bin", "python")
    
    # 2. Absolute path to the repo's offline script
    script_path = os.path.join(SAM_BODY4D_REPO_DIR, "scripts/offline_app.py") 
    
    # 3. Construct the command
    command = [
        python_path, 
        script_path, 
        "--input_video", video_path
    ]
    
    subprocess.run(command, cwd = SAM_BODY4D_REPO_DIR)


# Usage
# run_sam_body4d("input/my_video.mp4", "/home/user/anaconda3/envs/body4d")
```
