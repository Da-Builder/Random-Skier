# How to use code from a GitHub repo

## Submodule

It stores a pointer to a specific in the other repo. This is ideal when you might pull or push updates.

## Dependency (when the repo is designed to be a library)

Create a `requirements.txt` or `pyproject.toml` and install the repo directly via pip. 

## Docker (when the repo is designed to be an application)

In AI/ML this is the standard - it's good because if it works on your machine, it will work on a server or any machine without thm installing depednencies manually. I don't really know how it works.

## Side-by-side folders

You have two separate folders (your repo and the other repo) locally on your machine. Create one conda environment that works for both repos (just tailor it towards the other repo) this includes dependencies.
